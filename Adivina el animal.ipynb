```python
from openai import OpenAI

openai = OpenAI(api_key='')
```

```python
import random

ANIMALS = ['elefante', 'león', 'jirafa', 'hipopótamo', 'mono','jaguar','tucán','guacamayo','oso perezoso','gorila']

def get_clue():
    word = random.choice(ANIMALS)
    prompt = 'Adivina la palabra. Es un animal de la selva.'
    return prompt, word
```

```python
def check_answer(user_input, answer):
    return user_input.strip().lower() == answer.lower()
```

```python
def give_property(animal):
    try:
        response = openai.chat.completions.create(
            model='gpt-5.4-nano',
            messages=[{"role": "user", "content": f"Dame una característica de un {animal}, pero jamás digas el nombre del animal"}],
            max_completion_tokens=50
        )
        print("Pista:", response.choices[0].message.content.strip())
    except Exception as e:
        return "No pude generar una pista en este momento."
```

```python
def play_game():
    prompt, answer = get_clue()
    print(prompt)

    attempts = 0
    max_attempts = 5

    while attempts < max_attempts:
        user_input = input('Ingresa tu respuesta: ')

        if check_answer(user_input, answer):
            print(f'Correcto! La respuesta era: {answer}')
            return
        else:
            attempts += 1
            print('Incorrecto.')
            give_property(answer)

    print(f'Se acabaron los intentos. La respuesta era: {answer}')
```

```python
play_game()
```
